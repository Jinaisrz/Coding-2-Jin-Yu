```python
import os # Import the operating system module to operate files and directories.
files = os.listdir('YUJIN.coding2') # Gets the names of all files and folders in the specified folder 'YUJIN.coding2' and stores them in a list files.

import matplotlib.pyplot as plt 
import numpy as np
# Import the matplotlib and numpy modules for drawing and array manipulation.

files = [os.path.join('YUJIN.coding2', file_i)
 for file_i in os.listdir('YUJIN.coding2')
 if '.jpg' in file_i]
# Using list comprehensions, get the full path of all picture files with the suffix '.jpg' in the folder 'YUJIN.coding2' and store them in list Files.

imgs = [plt.imread(files[file_i])
        for file_i in range(50)]
# Read the first 50 images and store their pixel values in the list imgs. plt.imread() is a function in the matplotlib library for reading image files.

data = np.array(imgs) 
# Convert the list imgs to numpy array data to facilitate statistical manipulation of them.

mean_img = np.quantile(data,0.15,axis=0) 
# Calculate 90% of the quantiles of the array data, which will give you an average image. Where axis=0 means operation on the vertical axis (i.e., row) of the pixel, that is, statistical operation is performed on the corresponding pixel of all pictures.

plt.imshow(mean_img.astype(np.uint8))
# Display the average image using the imshow function of the matplotlib library. astype(np.uint8) converts the array type to an 8-bit unsigned integer type for easier display.
```
